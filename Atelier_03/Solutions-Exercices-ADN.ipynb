# Une chaine de caractère importante: l'ADN

```python
DNA = 'CCGGAAGAGCTTACTTAG'
```

## Premier exercice
Compter le nombre de nucléotides de chaque type dans une séquence d'ADN.

```python
for nuc in {'A','T','G','C'}:
    print(nuc, ':', sum([base==nuc for base in DNA]))
```

## Deuxème exercice
Faire une fonction qui calcule le taux de GC d'un brin d'ADN.

```python
GC = 100.*sum([base=='G' or base=='C' for base in DNA])/len(DNA)
print('GC =', GC)
```

```python
def GCcontent(DNA):
    return 100.*sum([base=='G' or base=='C' for base in DNA])/len(DNA)
```

```python
print('GC =', GCcontent(DNA))
```

Faire une fonction qui traduit un brin d'ARN en protéine.

On commence par extraire les codons:

```python
[DNA[i:i+3] for i in range(0, len(DNA), 3)]
```

```python
RNA = ''.join([basecomplement[base] for base in DNA[::-1]]).replace('T', 'U')
for c in [RNA[i:i+3] for i in range(0, len(RNA), 3)]:
    print(c, ':', map[c])
```
